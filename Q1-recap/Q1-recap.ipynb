```python
#############
# FUNCTIONS
#############
def custom_autopct_count(pct, allvals):
    absolute = int(np.round(pct/100.*np.sum(allvals)))
    # return f"{absolute:d}\n({pct:.1f}%)"
    return f"{absolute:d}"

def custom_autopct_percent(pct):
    return ('%1.1f%%' % pct) if pct > 5 else ''

def custom_label(names):
    labels = list(names)
    global wrapped_labels
    wrapped_labels = []
    for text in labels:
        wrapped_labels.append(textwrap.fill(text, width=16, break_long_words=False))
        
    return(wrapped_labels)

def custom_wedges():
    tempList = random.sample(clrs, 6)
    global colors
    colors = []
    for i in range(len(wrapped_labels)):
        if i <= 5:
            colors.append(tempList[i])
        elif i >= 6:
            colors.append(tempList[i-6])
    
    return(colors)

def custom_piechart():
    for pct, color in zip(pcts, pctscolors):
        pct.set_color(color)
        
    kw = dict(arrowprops=dict(arrowstyle="-"), zorder=0, va="center")
    for i, p in enumerate(wedges):
        ang = (p.theta2 - p.theta1)/2. + p.theta1
        y = np.sin(np.deg2rad(ang))
        x = np.cos(np.deg2rad(ang))
        horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
        connectionstyle = "angle,angleA=0,angleB={}".format(ang)
        kw["arrowprops"].update({"connectionstyle": connectionstyle})
        ax.annotate(wrapped_labels[i], xy=(x, y), xytext=(1.1*np.sign(x), 1.1*y), horizontalalignment=horizontalalignment, **kw)

    center_circle = plt.Circle((0, 0), 0.65, color="#E2EDFF")
    fig.gca().add_artist(center_circle)
    plt.setp(pcts, fontsize=32)
    fig.set_facecolor("#E2EDFF")
    plt.subplots_adjust(top=0.85)
    plt.text(0.02, 0.02, "@HANDMADEPHD", fontsize=42, transform=plt.gcf().transFigure)
    plt.text(0.7625, 0.02, "2023 Q1 REVIEW", fontsize=42, transform=plt.gcf().transFigure)

def secondsToTime(seconds):
    minutes, seconds = divmod(seconds, 60)
    hours, minutes = divmod(minutes, 60)
    return "%d:%02d" % (hours, minutes)
```

```python
##############################
# TIME PER COMPLETED SEWING
##############################
tab = df[(df["TASK"] != "KNIT")]
tab = tab.stb.freq(["PROJECT NAME", "TASK"], value="SECONDS")
tab = tab.iloc[:, 0:3]
tab = tab.pivot(index="PROJECT NAME", columns=["TASK"])
tab = tab.reset_index()
tab.fillna(0, inplace=True)
tab = pd.DataFrame(tab.to_records(index=False))
tab.columns = [list(tab.columns[0].split("'"))[1],
               list(tab.columns[1].split("'"))[3],
               list(tab.columns[2].split("'"))[3],
               list(tab.columns[3].split("'"))[3],
               list(tab.columns[4].split("'"))[3]]

for p in df["PROJECT NAME"].unique():
    tab.loc[tab["PROJECT NAME"]==p, "DATE"] = df.loc[df["PROJECT NAME"]==p, "DATE"].max()

tab.index = tab["DATE"]
tab = tab.sort_index()
tab = pd.DataFrame(tab.to_records(index=False))

for p in df["PROJECT NAME"].unique():
  strdate = df.loc[(df["TASK"]=="COMPLETE") & (df["PROJECT NAME"]==p), "DATE"].max()
  if pd.isna(strdate) == True:
    tab.drop(tab[tab["PROJECT NAME"]==p].index, inplace=True)
  if pd.isna(strdate) == False:
    if (df.loc[(df["TASK"]=="COMPLETE") & (df["PROJECT NAME"]==p), "DATE"].max().strftime("%Y") > "2023") == True:
        tab.drop(tab[tab["PROJECT NAME"]==p].index, inplace=True)

tab = tab.reset_index()
tab['TOTAL'] = tab.sum(axis=1)
tab_average = tab['TOTAL'].sum() / len(tab)
k = (secondsToTime(tab_average))
```

```python
###############################
# TIME PER COMPLETED KNITTING
###############################
tab = df[(df["TASK"] == "KNIT")]
tab = tab.stb.freq(["PROJECT NAME", "TASK"], value="SECONDS")
tab = tab.iloc[:, 0:3]
tab = tab.pivot(index="PROJECT NAME", columns=["TASK"])
tab = tab.reset_index()
tab.fillna(0, inplace=True)
tab = pd.DataFrame(tab.to_records(index=False))
tab.columns = [list(tab.columns[0].split("'"))[1],
               list(tab.columns[1].split("'"))[3]]

for p in df["PROJECT NAME"].unique():
    tab.loc[tab["PROJECT NAME"]==p, "DATE"] = df.loc[df["PROJECT NAME"]==p, "DATE"].max()

tab.index = tab["DATE"]
tab = tab.sort_index()
tab = pd.DataFrame(tab.to_records(index=False))

for p in df["PROJECT NAME"].unique():
  strdate = df.loc[(df["TASK"]=="COMPLETE") & (df["PROJECT NAME"]==p), "DATE"].max()
  if pd.isna(strdate) == True:
    tab.drop(tab[tab["PROJECT NAME"]==p].index, inplace=True)
  if pd.isna(strdate) == False:
    if (df.loc[(df["TASK"]=="COMPLETE") & (df["PROJECT NAME"]==p), "DATE"].max().strftime("%Y") > "2023") == True:
        tab.drop(tab[tab["PROJECT NAME"]==p].index, inplace=True)

tab = tab.reset_index()
tab['TOTAL'] = tab.sum(axis=1)
tab_average = tab['TOTAL'].sum() / len(tab)
l = (secondsToTime(tab_average))
```
